```python
print("🚦 Starting SUMO simulation...")

import os
os.environ['SUMO_GUI'] = 'true'   # set to 'false' for headless mode

from simulation.simulation_manager import SimulationManager

try:
    mgr = SimulationManager()
    results = mgr.run()
    print("✅ Simulation completed")
    print('Summary:', results['summary'])
except Exception as e:
    print("❌ Simulation failed:", e)
```

```python
print("📊 Plotting results...")

import pandas as pd
import matplotlib.pyplot as plt

history = results['history']
df = pd.DataFrame(history)

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Simulation Results', fontsize=14)

df.plot(x='step', y='avg_waiting_time', ax=axes[0,0], title='Avg Waiting Time (s)', legend=False)
df.plot(x='step', y='total_queue',       ax=axes[0,1], title='Total Queue', legend=False)
df.plot(x='step', y='avg_speed_kmh',     ax=axes[1,0], title='Avg Speed (km/h)', legend=False)
df.plot(x='step', y='throughput',        ax=axes[1,1], title='Throughput (veh/step)', legend=False)

plt.tight_layout()
plt.savefig('logs/simulation_results.png', dpi=120)
plt.show()

print('✅ Chart saved to logs/simulation_results.png')
```
